<h1><center>LINE CHARTS</center></h1>

Line charts are pretty much useful and very common in practice. Line charts are the way to depict relations b/w two continuous variables and hence they are bivariate i.e you need at least 2 variables to plot a line chart.

**Advantages :**
1. Easy to interpret the trends over time and also helps to extrapolate the data.
2. A-line break can also depict the presence of missing data.
3. We can also estimate the correlation value by seeing the steepness in the line.
**Disadvantages:**
1. Though line charts are easy to interpret but plotting two line charts over the same figure can make it difficult to compare the results.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'E:\Downloads\auto-mpg.csv', na_values = ['?'])
df.head()
```

```python
df.isna().sum()
```

```python
df['horsepower'].fillna(df['horsepower'].median(), inplace = True)
```

```python
df.sort_values(by = 'mpg', inplace = True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
plt.plot(df['mpg'], df['displacement'])
plt.show()
```

```python
plt.plot(df['mpg'], df['displacement'])
plt.xlabel('MPG', size = 15, fontweight = 'bold')
plt.ylabel('Displacement', size = 15, fontweight = 'bold')
plt.title('Trend b/w mpg & displacement', size = 20, fontweight = 'bold')
plt.show()
```

```python
plt.plot(df['mpg'], df['displacement'], color = 'k')
plt.xlabel('MPG', size = 15, fontweight = 'bold')
plt.ylabel('Displacement', size = 15, fontweight = 'bold')
plt.title('Trend b/w mpg & displacement', size = 20, fontweight = 'bold')
#plt.grid()
plt.show()
```

```python
plt.plot(df['mpg'], df['displacement'], color = 'k',
        linestyle = '-.')
plt.xlabel('MPG', size = 15, fontweight = 'bold')
plt.ylabel('Displacement', size = 15, fontweight = 'bold')
plt.title('Trend b/w mpg & displacement', size = 20, fontweight = 'bold')
plt.grid()
plt.show()
```

```python
plt.plot(df['mpg'], df['displacement'], label = 'displacement', color ='g')
plt.plot(df['mpg'], df['horsepower'], label = 'horsepower', color ='r')
plt.xlabel('MPG')
plt.ylabel('Displacement / Horsepower')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (8,  6))
plt.plot(df['mpg'], df['displacement'], label = 'displacement', color ='g')
plt.plot(df['mpg'], df['horsepower'], label = 'horsepower', color ='r')
plt.xlabel('MPG')
plt.ylabel('Displacement / Horsepower')
plt.legend()
plt.grid()
plt.title('Comparison b/w displacement & horsepower')
plt.savefig(r'E:\Downloads\linechart.png')
```

```python
plt.plot(df['mpg'], df['displacement'], color = 'k',
        linestyle = '-.', marker = '+', markersize = 10)
plt.xlabel('MPG', size = 15, fontweight = 'bold')
plt.ylabel('Displacement', size = 15, fontweight = 'bold')
plt.title('Trend b/w mpg & displacement', size = 20, fontweight = 'bold')
plt.grid()
plt.show()
```

```python
plt.plot(df['mpg'], df['displacement'], linestyle = ':',
        linewidth = 3)
plt.xlabel('MPG')
plt.ylabel('Displacement')
plt.show()
```
